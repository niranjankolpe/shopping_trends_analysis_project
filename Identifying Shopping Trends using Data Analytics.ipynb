```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "shopping_trends.csv"
df = pd.read_csv(file_path)
```

```python
# Basic dataset overview
print("Dataset Info:\n")
df.info()
print("\nBasic Statistics:\n", df.describe())
```

## What is the gender distribution of customers?

```python
# Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, hue='Gender', palette='coolwarm', legend=False)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## What is the age distribution of customers?

```python
# Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

## What are the most purchased product categories?

```python
# Popular Product Categories
plt.figure(figsize=(10, 5))
sns.countplot(y='Category', data=df, order=df['Category'].value_counts().index, hue='Category', palette='viridis', legend=False)
plt.title('Most Purchased Product Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

## How is the distribution of purchase amounts?

```python
# Purchase Amount Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Purchase Amount (USD)'], color='green')
plt.title('Purchase Amount Distribution')
plt.xlabel('Purchase Amount (USD)')
plt.show()
```

## How do purchases vary by season?

```python
# Seasonal Shopping Trends
plt.figure(figsize=(8, 5))
sns.countplot(x='Season', data=df, order=df['Season'].value_counts().index, hue='Season', palette='pastel', legend=False)
plt.title('Purchases by Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

## How do discounts impact purchases?

```python
# Effect of Discount on Purchases
plt.figure(figsize=(6, 4))
sns.countplot(x='Discount Applied', data=df, hue='Discount Applied', palette='Set1', legend=False)
plt.title('Discount Impact on Purchases')
plt.xlabel('Discount Applied')
plt.ylabel('Count')
plt.show()
```

## Is there a correlation between previous purchases and purchase amount?

```python
# Correlation between previous purchases and purchase amount
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Previous Purchases', y='Purchase Amount (USD)', data=df, alpha=0.6)
plt.title('Previous Purchases vs Purchase Amount')
plt.xlabel('Previous Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

## What are the most common payment methods used by customers?

```python
# Payment Methods Used
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Payment Method'].value_counts().index, y=df['Payment Method'].value_counts().values)
plt.title('Most Common Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
